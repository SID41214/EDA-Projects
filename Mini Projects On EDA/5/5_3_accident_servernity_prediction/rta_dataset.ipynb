# Road traffic accidents (RTA)

# 1). problem statement:



## Data Description

The data set has been prepared from manual records of road traffic accident of the year 2017-20. All the sensitive information have been excluded during data encoding and finally it has 32 features and 12316 instances of the accident. Then it is preprocessed and for identification of major causes of the accident by analyzing it using different machine learning classification algorithms algorithms. RTA Dataset.csv is the dataset before preprocessing and cleaned.csv is the preprocessed dataset.


# Problem Statement:

The goal of this project is to predict the **severity of road traffic accidents** using a variety of factors, including driver characteristics, vehicle details, road conditions, and environmental factors. The severity of an accident is classified into categories such as **no injury**, **serious injury**, or **fatality**. By identifying the factors that influence the severity of an accident, we aim to contribute valuable insights to improve road safety, assist in accident prevention, and support policy-making.

### Dataset Details:
The dataset contains the following columns, which are used as features to predict the target variable:

- **Time**: Time of day when the accident occurred.
- **Day_of_week**: Day of the week when the accident occurred.
- **Age_band_of_driver**: Age group of the driver.
- **Sex_of_driver**: Gender of the driver.
- **Educational_level**: Education level of the driver.
- **Vehicle_driver_relation**: Relationship between the driver and the vehicle (e.g., owner, rental).
- **Driving_experience**: Number of years the driver has been driving.
- **Type_of_vehicle**: Type or category of the vehicle involved in the accident.
- **Owner_of_vehicle**: Ownership status of the vehicle.
- **Service_year_of_vehicle**: Age of the vehicle service (year of service).
- **Defect_of_vehicle**: Whether the vehicle had any defect at the time of the accident.
- **Area_accident_occured**: Geographical area where the accident took place.
- **Lanes_or_Medians**: Presence of lanes or medians on the road.
- **Road_allignment**: Road alignment type (e.g., straight, curved).
- **Types_of_Junction**: Type of road junction involved.
- **Road_surface_type**: Surface material of the road (e.g., asphalt, gravel).
- **Road_surface_conditions**: Condition of the road surface (e.g., wet, dry).
- **Light_conditions**: Visibility or lighting conditions at the time of the accident (e.g., daylight, night).
- **Weather_conditions**: Weather conditions at the time of the accident (e.g., clear, rainy, foggy).
- **Type_of_collision**: The type of collision (e.g., head-on, side swipe).
- **Number_of_vehicles_involved**: Number of vehicles involved in the accident.
- **Number_of_casualties**: Number of people injured in the accident.
- **Vehicle_movement**: Movement of the vehicle at the time of the accident (e.g., moving, stationary).
- **Casualty_class**: Class of casualties involved (e.g., pedestrian, driver).
- **Sex_of_casualty**: Gender of the casualty.
- **Age_band_of_casualty**: Age group of the casualty.
- **Casualty_severity**: Severity of the casualty (e.g., killed, seriously injured).
- **Work_of_casualty**: Occupation of the casualty (if available).
- **Fitness_of_casualty**: Fitness status of the casualty (e.g., under the influence, fit).
- **Pedestrian_movement**: Movement of pedestrians at the time of the accident (if applicable).
- **Cause_of_accident**: Main cause of the accident (e.g., driver error, vehicle defect).
  
### Target Variable:
- **Accident_severity**: The severity of the accident, which will be the target variable for prediction. It is typically a categorical variable with values like:
  - **0**: No injury (or minor injury)
  - **1**: Serious injury
  - **2**: Fatality

### Model:
For this classification problem, machine learning models such as **Decision Trees**, **Random Forests**, **Logistic Regression**, or **Gradient Boosting** can be used. The model will predict the **Accident_severity** based on the given features.

Evaluation metrics for model performance would include **accuracy**, **precision**, **recall**, and **F1-score**, with a focus on correctly identifying more severe accidents (fatalities or serious injuries).

This analysis will help identify significant risk factors influencing accident severity and support better road safety initiatives.

```python
def get_time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 24:
        return 'Evening'
    else:
        return 'Night'

df1['Time_of_Day'] = df1['Time'].apply(get_time_of_day)
df1 = pd.get_dummies(df1, columns=['Time_of_Day'], prefix=['Time_of_Day'])
```

# modeling

```python
# Import the necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
```

```python
y=y.astype(int)
x_train,x_test,y_train,y_test=train_test_split(df1_pca,y,test_size=0.2,random_state=42)
```

```python

# List of classification models to apply
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Support Vector Classifier": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Neural Network": MLPClassifier(max_iter=1000)
}

# Function to evaluate model performance
def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average="weighted")
    recall = recall_score(y_true, y_pred, average="weighted")
    f1 = f1_score(y_true, y_pred, average="weighted")
    return accuracy, precision, recall, f1
```

```python

# Dictionary to store the results
results = {}

# Apply each model and compute metrics
for name, model in models.items():
    if name in ["Support Vector Classifier", "Neural Network"]:  # Models requiring scaling
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
    else:
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
    # Compute evaluation metrics
    accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred)
    results[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1
    }

```

```python
results_df = pd.DataFrame(results).T  # Transpose for better readability
results_df.index.name = "Model"
print(results_df)
```
